```python
! pip install numpy scipy
import numpy as np
from scipy.optimize import linprog
```

```python
problem_type = None
num_vars = 0
num_constraints = 0

c = []
A_ub , b_ub = [], []
A_eq , b_eq = [], []
```

```python
def solve():
    global problem_type, num_vars, num_constraints
    global c, A_ub, b_ub, A_eq, b_eq

    problem_type = input("Enter problem type (min/max): ").strip().lower()
    num_vars = int(input("Enter number of variables: ").strip())
    num_constraints = int(input("Enter number of constraints: ").strip())

    for i in range(num_vars):
        coeff = float(input(f"Enter coefficient for variable x{i+1} in objective function: ").strip())
        c.append(coeff)

    for i in range(num_constraints):
        constraint_type = input(f"Enter constraint type for constraint {i+1} (<=, >=, =): ").strip()
        coeffs = []
        for j in range(num_vars):
            coeff = float(input(f"Enter coefficient for variable x{j+1} in constraint {i+1}: ").strip())
            coeffs.append(coeff)
        rhs = float(input(f"Enter right-hand side value for constraint {i+1}: ").strip())

        if constraint_type == "<=":
            A_ub.append(coeffs)
            b_ub.append(rhs)
        elif constraint_type == ">=":
            A_ub.append([-x for x in coeffs])
            b_ub.append(-rhs)
        elif constraint_type == "=":
            A_eq.append(coeffs)
            b_eq.append(rhs)

    A_ub = np.array(A_ub) if A_ub else None
    b_ub = np.array(b_ub) if b_ub else None
    A_eq = np.array(A_eq) if A_eq else None
    b_eq = np.array(b_eq) if b_eq else None

    if problem_type == "max":
        c = [-x for x in c]
    res = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq, method='highs')
    if res.success:
        if problem_type == "max":
            print("Optimal value (max):", -res.fun)
        else:
            print("Optimal value (min):", res.fun)

        print("Variable values:", res.x)
    else:
        print("No solution found.")
```

```python
c = []
A_ub , b_ub = [], []
A_eq , b_eq = [], []

solve()
```
